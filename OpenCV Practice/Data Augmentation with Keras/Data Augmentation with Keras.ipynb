```python
from numpy import expand_dims
from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img = load_img('girl.jpg')
img = img_to_array(img)
data = expand_dims(img, 0)

# 1. Zoom image from 0.5x to 2.0x
myImageGen = ImageDataGenerator(zoom_range = [0.5, 2.0])
gen = myImageGen.flow(data, batch_size = 1)
for i in range(9):
    plt.subplot(3, 3, 1+i)
    myBatch = gen.next()
    image = myBatch[0].astype('uint8')
    plt.imshow(image)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    cv2.imwrite('aug/zoom' + str(i+1) + '.png', image)
plt.show()
```

```python
# 2. Flip image vertical and horizontal
myImageGen = ImageDataGenerator(horizontal_flip = True, vertical_flip = True)
gen = myImageGen.flow(data, batch_size = 1)
for i in range(9):
    plt.subplot(3, 3, 1+i)
    myBatch = gen.next()
    image = myBatch[0].astype('uint8')
    plt.imshow(image)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    cv2.imwrite('aug/flip' + str(i+1) + '.png', image)
plt.show()
```

```python
# 3. Shift image up, down, left, right
myImageGen = ImageDataGenerator(width_shift_range = [-150, 150]) # oder width_shift_range = 0.25
gen = myImageGen.flow(data, batch_size = 1)
for i in range(9):
    plt.subplot(3, 3, 1+i)
    myBatch = gen.next()
    image = myBatch[0].astype('uint8')
    plt.imshow(image)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    cv2.imwrite('aug/shift_width' + str(i+1) + '.png', image)
plt.show()

myImageGen = ImageDataGenerator(height_shift_range = [-200, 200])
gen = myImageGen.flow(data, batch_size = 1)
for i in range(9):
    plt.subplot(3, 3, 1+i)
    myBatch = gen.next()
    image = myBatch[0].astype('uint8')
    plt.imshow(image)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    cv2.imwrite('aug/shift_height' + str(i+1) + '.png', image)
plt.show()
```

```python
# 4. Adjust image's brightness
myImageGen = ImageDataGenerator(brightness_range = [0.2, 2.0])
gen = myImageGen.flow(data, batch_size = 1)
for i in range(9):
    plt.subplot(3, 3, 1+i)
    myBatch = gen.next()
    image = myBatch[0].astype('uint8')
    plt.imshow(image)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    cv2.imwrite('aug/bright' + str(i+1) + '.png', image)
plt.show()
```

```python
# 5. Rotate image
myImageGen = ImageDataGenerator(rotation_range = 45)
gen = myImageGen.flow(data, batch_size = 1)
for i in range(9):
    plt.subplot(3, 3, 1+i)
    myBatch = gen.next()
    image = myBatch[0].astype('uint8')
    plt.imshow(image)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    cv2.imwrite('aug/rotate' + str(i+1) + '.png', image)
plt.show()
```

```python
# 6. Shear image
myImageGen = ImageDataGenerator(shear_range = 45)
gen = myImageGen.flow(data, batch_size = 1)
for i in range(9):
    plt.subplot(3, 3, i+1)
    myBatch = gen.next()
    image = myBatch[0].astype('uint8')
    plt.imshow(image)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    cv2.imwrite('aug/shear' + str(i+1) + '.png', image)
plt.show()
```
